```python

def aggregate_player_stats_explicit(group):
    # Initialize the dictionary to store aggregated results
    aggregated_data = {}

    # Explicitly handle the weighted average calculation for specified columns
    weighted_columns = ['FT%', 'ORtg', '3P%', 'PORPAG', 'Adj_O_Eff', 'Mid%', 'O_BPM', 'PTS_PG', '3PT/100', 'Mid_Missed', 'TS%', 'USG']
    for col in weighted_columns:
        if len(group) > 1:
            # If there are multiple seasons, add the last season's value twice, sum, then divide by len(group) + 1
            weighted_average = (group[col].sum() + group[col].iloc[0]) / (len(group) + 1)
            #weighted_average = group[col].sum()/len(group)
        else:
            # If there's only one season, just use its value as is
            weighted_average = group[col].iloc[0]
        aggregated_data[col] = weighted_average

    # Summation for cumulative statistic columns
    sum_columns = ['GP', 'FTM', 'Mid_Made', '3PA', 'Mid_Missed', '3P']
    for col in sum_columns:
        aggregated_data[col] = group[col].sum()

    # Handling uniform columns, assuming the most recent season is the last entry
    uniform_columns = ['Recruiting_Rank', 'Class', 'Year', 'Height']
    for col in uniform_columns:
        aggregated_data[col] = group[col].iloc[0]

    return pd.Series(aggregated_data)

# Apply the function across players
aggregated_ncaa_df = ncaa_filtered_columns_df.groupby('Player').apply(aggregate_player_stats_explicit).reset_index()
pd.set_option("display.max_rows", None)
aggregated_ncaa_df
```
